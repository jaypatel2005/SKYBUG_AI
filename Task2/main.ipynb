```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def cnvrt(obj):
    l = []
    for item in ast.literal_eval(obj):
        l.append(item['name'])
    
    return l
```

```python
df['genres'] = df['genres'].apply(cnvrt)
```

```python
df['keywords'] = df['keywords'].apply(cnvrt)
```

```python
def cnvrt3(obj):
    l = []
    counter = 0
    for item in ast.literal_eval(obj):
        if counter != 3:
            l.append(item['name'])
        else:
            break
        counter += 1
    
    return np.array(l)
```

```python
cnvrt3(df.iloc[0]['cast'])
```

```python
df['cast'] = df['cast'].apply(cnvrt3)
```

```python
def get_director(obj):
    l = []
    
    for item in ast.literal_eval(obj):
        if item['job'] == 'Director':
            l.append(item['name'])
            break
    
    return l
```

```python
get_director(df.iloc[0]['crew'])
```

```python
df['Director'] = df['crew'].apply(get_director)
```

```python
df.isnull().sum()
```

```python
df.sample(10)
```

```python
def collasp(l):
    l1 = []
    
    for item in l:
        l1.append(item.replace(" ", ""))
        
    return l1
```

```python
collasp(["Jennifer Lawrence"])
```

```python
df['genres'] = df['genres'].apply(collasp)
df['cast'] = df['cast'].apply(collasp)
df['Director'] = df['Director'].apply(collasp)
df['keywords'] = df['keywords'].apply(collasp)
```

```python
df.sample(10)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
np.array(df['Director'])
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['Director']
```

```python
new_df = df[['title', 'movie_id', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new_df['title'] = new_df['title'].apply(lambda x: x.lower())
```

```python
new_df['title']
```

```python
new_df.iloc[10]['title']
```

```python
def recommand(movie, n_movies=5):
    try:
        idx = new_df[new_df['title'] == movie.lower()].index[0]
    except IndexError:
        print("The movie name you entered is not available in our dataset")
        return
    except Exception as e:
        print(e)
        return
    
    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x:x[1])
    
    recommanded = []
    
    for i in range(1, n_movies+1):
        recommanded.append(new_df.iloc[distances[i][0]]['title'])
    
    return recommanded
```

```python
recommand('Harry Potter and the Half-Blood Prince', 10)
```
